```python
class Pesquisa:
    
    data_hoje = datetime.datetime.today().strftime('%d/%m/%Y')
    
    lista = {'Data':[], 'Estabelecimento':[], 'Produto Scrap':[], 'Marca':[],'Preço':[]}
    
    marcas = ['Cottonbaby', 'Intimus', 'Plenitud', 'Always', 'Sempre Livre', 'Mili']
    
    def __init__(self, estabelecimento, url, tags, tag_nome, classe_nome, tag_preço, classe_preço):
        self.estabelecimento = estabelecimento
        self.url = url
        self.tags = tags
        self.tag_nome = tag_nome
        self.classe_nome = classe_nome
        self.tag_preço = tag_preço
        self.classe_preço = classe_preço
    
    def pesquisar(self):
        
        # Pesquisa pelas palavras dentro da lista
        for termo in self.tags:
            
            # Formata a url com a palavra chave
            url_nova = self.url.format(termo)
            
            # Faz a request pro site usando a url formatada, o .text retorna o texto contido
            html_text = requests.get(url_nova).text
            
            # Aplica o parser 'lxml' para 'correr' pelo esqueleto da page
            soup = BeautifulSoup(html_text, 'lxml')
            
            # Busca pelo nome do item
            html_item = soup.find_all(self.tag_nome, class_ = self.classe_nome)
            
            # Busca pelo preço do item
            html_preço = soup.find_all(self.tag_preço, class_ = self.classe_preço)

            # Faz os loops de escrita do item
            for produto in html_item:
                self.lista['Data'].append(self.data_hoje)
                self.lista['Estabelecimento'].append(self.estabelecimento)
                self.lista['Produto Scrap'].append(produto.text)
                self.lista['Marca'].append('nulo')


            for preço in html_preço:
                self.lista['Preço'].append(preço.text.replace('\n',''))  
        
        # Coloca a marca lida da string do Scrap
        for produto in self.lista['Produto Scrap']:
            
            produto_split = produto.split() #split as palavras da linha
            
            if i == self.marcas in produto_split:
                self.lista['Marca'].replace('nulo',i)
                
```

```python
# URL: o {} é a palavra chave, obtida direto da URL da página
url = 'https://www.drogariacatarinense.com.br/{}'

# Palavras para a Busca
hastes = ['hastes%20flexiveis']
absorventes = ['absorvente%20com%20abas%208%20unidades','absorvente%20noturno']
algodao = ['algodao%20bola','algodao%20rolo', 'algodao%20disco']
curativo = ['curativo%2015','curativo%2030']

palavras_chave = [hastes, absorventes, algodao, curativo]

# Estabelecimento
estabelecimento = 'Drogaria Catarinense'
```

```python
# Loop de busca das palavras chave

# Faz o loop na lista de tags


for tag in palavras_chave:
    for termo in tag:
    
        url_nova = url.format(termo) # Url formatada com a palavra chave

        html_text = requests.get(url_nova).text #string para botar a URL da pagina dentro do '' // sem o .text no fim, eu recebo apenas a mensagem que o request de acesso obteve sucesso "código 200"

        soup = BeautifulSoup(html_text, 'lxml') #abre a instancia carregada do BSoup para ler a variável e usa o parser 'lxml'.

        html_item = soup.find_all('h3', class_ = "product-item__title") #_next(class_ = ["product-item__title", 'product-item__price']) #cria a variável html_cards que usará uma função do soup de encontrar os parametros

        html_preço = soup.find_all('div', class_ = 'product-item__best-price')

        for produto in html_item:
            lista['Data'].append(data_hoje)
            lista['Estabelecimento'].append(estabelecimento)
            lista['Produto Scrap'].append(produto.text)
            lista['Marca'].append('nulo')

        for preço in html_preço:
            lista['Preço'].append(preço.text.replace('\n',''))
```
